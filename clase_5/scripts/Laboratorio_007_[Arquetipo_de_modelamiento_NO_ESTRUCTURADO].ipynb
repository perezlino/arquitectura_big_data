### Tensorización de las imagenes [Forma correcta]

```python
# Vamos a importar una librería que está especializada en convertir imágenes en representación tensorial. No nos va a importar mucho 
# realmente que hace la librería, solamente quiero que vean cómo es que existen utilitarios para procesamiento de datos no estructurados. 
from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
# Por ejemplo, voy a decir lo siguiente: quiero crear un tensor que contenga todas las imágenes tensorizadas, si tenemos 10.000 imágenes, 
# este tensor va a tener 10.000 tensores, cada uno con 3 canales de colores. Entonces, desde esta función utilitaria que tensoriza imágenes 
# dentro de un directorio, abrimos y cerramos paréntesis vamos a configurar lo siguiente: en primer lugar, en donde están las imágenes a 
# tensorizar es dentro de LANDING, en el servidor de imágenes, en la ruta de los ANIMALES, además, son imágenes que tienen un color RGB, 
# tienen 3 canales de colores, porque, podría darse el caso en que te den imágenes en blanco y negro, por ejemplo, radiografías, entonces, 
# eso sería escala de grises, sería otra historia. Ahora, también tenemos el problema de que los tensores tienen que tener imágenes de un 
# mismo tamaño, eso lo podemos hacer desde esta librería y le decimos: “ … oye y vamos a redimensionar todas las imágenes, para que todas 
# midan 32 por 32 pixeles. Pero ¿cuál es el problema con redimensionar las imágenes? que, por ejemplo, las imágenes que son rectangulares 
# las estamos volviendo cuadradas y se van a distorsionar. Hay que utilizar un algoritmo de INTERPOLACIÓN para que esa distorsión sea mínima, 
# por ejemplo, existe el algoritmo bilineal. Ahora, otro punto importante, ojo, estamos explicando a nivel conceptual, si esto fuese una clase 
# de ciencia de datos ya iríamos con mayor detalle, porque, no es tan simple, solo entendamos a nivel conceptual. Otra variable y ¿qué es lo 
# que pasa con la tensorización de las imágenes? potencialmente consumen mucha memoria RAM de la infraestructura y, por ejemplo, si tenemos 
# un dataset de 100.000 imágenes de 32 POR 32 PÍXELES, no recuerdo el cálculo exacto pero eso aproximadamente sale que se necesitarían 40 GB 
# de memoria RAM para tensorizar esas 100.000 imágenes. ¿Qué es lo que podríamos hacer? en lugar de pasarle las 100.000 imágenes directamente 
# y que se ocupen 40 GB de golpe podríamos pasar en vez de 100.000, le pasamos de 10.000 y luego otras 10.000, por lo que, ya no serán los 
# 40 GB, como le pasamos de 10.000 en 10.000 ahora solamente se utilizarán 4 GB, se harán 10 iteraciones cada iteración solo ocupa 4 GB de RAM 
# y ya no se acumularán 40 GB de RAM. Eso es lo que se llama el TAMAÑO DE BATCH, de cuanto en cuanto queremos ir pasando las imágenes para 
# evitar que de un solo golpe se llene toda la RAM y se tensorize todo de una vez. Ese es el BATCH SIZE. Acá define, por ejemplo, si son 
# 100.000 imágenes podremos pasarle de 10.000 (10.000 y luego tomara otras 10.000). Nosotros tenemos menor cantidad de imágenes, por tanto, 
# voy a pasar de 10 en 10. Ahora, aquí vienen detalles técnicos que ya vienen del mundo de la ciencia de datos, por ejemplo, ¿para qué se 
# utilizan estas imágenes? podríamos construir un modelo de deep learning que haga reconocimiento de patrones visuales para que aprenda qué 
# significa aves, qué significa perros y qué significa gatos, porque, tenemos un dataset de aves, perros y gatos. Por ejemplo, las aves ¿que 
# tienen? tienen alas, tienen cierta forma de patas, su boca tiene forma de un triángulo, entonces ,digamos que es búsqueda de patrones. 
# Existen 3 tipos de imágenes que tenemos: aves, gatos y perros, eso en el mundo de la ciencia de datos se llama LABELS. Acá por ejemplo, le 
# decimos: “ … oye, quiero que infieras los nombres de los LABELS desde los directorios que los contienen … “. Recuerden que dentro del 
# LANDING, en el servidor de la fuente de imágenes (SERVER_IMG) tenemos para ANIMALES en la partición correspondiente (/2018-04-19), AVE, GATO 
# Y PERRO. El tensor debe tener al costado, se tensoriza y tiene que tener un número que diga si es un ave, un gato o un perro. Se va a colocar 
# en orden alfabético, AVE va a ser el número 0, GATO va a ser el número 1 y PERRO va a ser el número 2. Esto es lo que se llama inferir los 
# LABELS en función del subdirectorio en donde están las imágenes, ese va a ser el label asociado pero se van a colocar en forma de números, 
# entonces, le decimos: “ … cada imagen tensorizada tiene un label asociado: 0, 1 y 2 … “ ¿qué significa 0, 1 y 2? en orden alfabético, ave, 
# gato y perro. Y adicionalmente eso es LABELS queremos que se coloquen como 0, 1 y 2, para que no diga: ave, gato y perro, si no, 0, 1 y 2. 
# Entonces, cada tensor de cada imagen tiene asociado un número 0, 1 y 2 que indicará si es ave gato o perro. Por último, tenemos 100 imágenes 
# y se van a tensorizar de 10 en 10 para no llenar la memoria RAM. Como vamos a ir sacando las imágenes de 10 en 10, la extracción de esas 
# 10 imágenes se va a hacer de manera aleatoria, puede que saque 3 de aves, 2 de gatos y 5 de perros, ya va a ser completamente aleatoria la 
# tensorización. Ejecutamos esto y listo. 
```

```python
tensor = image_dataset_from_directory(
    "/content/drive/MyDrive/DATALAKE/LANDING/SERVER_IMG/ANIMALES/2018-04-19",
    color_mode="rgb",
    image_size=(32, 32),
    interpolation='bilinear',
    batch_size=10,
    labels="inferred",
    label_mode="int",
    shuffle=True
)
```

```python
# Hasta aquí entendimos la explicación conceptual, pero, con el código que utilizaremos ya es donde salimos de explicaciones conceptuales. 
# En el código anterior lo único que hemos hecho es configurar cómo va a ser este tensor, pero todavía no se ha tensorizado nada. Para eso 
# tendríamos que tener una clase de 3 horas de un curso de ciencia de datos para explicar esta parte de la tensorización por partes, no es 
# tan simple como configurar, hay que codificar algunas cosas, de hecho hay que colocar metadata y hacer ya cosas cada vez más complejas. 
# Un punto importante aquí es que generalmente las implementaciones estructuradas de los datos no estructurados lo tiene que hacer ya un 
# especialista, no lo puede hacer un Data engineer genérico, tiene que hacerlo un Data engineer especializado en el procesamiento de ese 
# tipo de datos no estructurados. Para poder explicarles este código, por ejemplo, requeriríamos literalmente, en el curso de ciencia de 
# datos en entornos de Big data que dicto, esto nos va a tomar 3 horas para poder entenderlo y hay cosas previas. Por ahora, va a ser simple, 
# es solamente una función utilitaria en donde le pasamos la configuración del tensor y la ruta en donde queremos almacenar el tensor, así 
# que, esta es una función utilitaria que en una clase de ciencia de datos tardaremos 3 horas en explicarlo. 
```

```python
# Importamos la función que categoriza
from tensorflow.keras.utils import to_categorical

# Función utilitaria para tensorización
def almacenar_en_disco_como_tensores(tensor, ruta):
    # Iteramos los tensores leidos
    index = 0
    for tensor_features, tensor_labels in iter(tensor):
        # Convertimos los features a array
        tensor_features_numpy = tensor_features.numpy()

        # Convertimos los labels a categóricos
        tensor_labels_categorico = to_categorical(tensor_labels, num_classes = 3)

        # Colocamos los datos en un diccionario
        dataset = {}
        dataset['features'] = tensor_features_numpy
        dataset['labels'] = tensor_labels_categorico
        dataset['metadata'] = tensor.class_names

        # Guardamos los datos tensoriales en un archivo
        dataset_numpy = np.array(dataset)
        np.savez(ruta+'/tensor_' + str(index), dataset_numpy)

        index = index + 1
```

### Almacenar tensores

```python
# Luego, almacenaremos en el disco duro esa configuración de las 100 imágenes encontradas dentro de UNIVERSAL en la carpeta de ANIMALES 
# en la partición correspondiente al día de “2018-04-19”. Ejecutamos, y vamos a esperar a que se creen los archivos tensorizados. 
almacenar_en_disco_como_tensores(tensor, '/content/drive/MyDrive/DATALAKE/UNIVERSAL/ANIMALES/2018-04-19')
```
